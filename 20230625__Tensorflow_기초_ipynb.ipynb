```python
import tensorflow as tf

# Define the tensors
tensor1 = tf.constant([[1, 2], [3, 4]])  # Shape: (2, 2)
tensor2 = tf.constant([[5, 6], [7, 8]])  # Shape: (2, 2)

# Perform matrix multiplication
result = tf.matmul(tensor1, tensor2)

# Print the result
print(result)
```
