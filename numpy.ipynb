```python
matrix_1 = [[1, 2],[3, 4]]
matrix_2 = [[5, 6],[7, 8]]
```

```python
matrix_result= []
for i in range(len(matrix_1)):
    tmp = []
    for j in range(len(matrix_2)):
        tmp.append(matrix_1[i][j]+matrix_2[i][j])
    matrix_result.append(tmp)

print(matrix_result)
```

```python
age = 9

def Age_Range(age) :
  if age // 10 == 0 :
    return "유아"
  elif age // 10 == 1 :
    return "10대"
  elif age // 10 == 2 :
    return "20대"
  elif age // 10 == 3 :
    return "30대"
  elif age // 10 == 4 :
    return "40대"
  elif age // 10 == 5 :
    return "50대"
  elif age // 10 == 6 :
    return "60대"
  else :
    return "노인"
```

```python
df['연령대'] = df['Age'].apply(Age_Range)
```
