# 1) 결측지처리를해주세요.(age-평균값적용,Cabin-N으로,Embarked-N)

```python
import pandas as pd

titanic_df = pd.read_csv('titanic_train.csv')
```

```python
# Age 열의 결측치를 평균값으로 채우기
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)

# Cabin 열의 결측치를 'N'으로 채우기
titanic_df['Cabin'].fillna('N', inplace=True)

# Embarked 열의 결측치를 'N'으로 채우기
titanic_df['Embarked'].fillna('N', inplace=True)
```

```python
titanic_df
```

# 2) 성별(Sex)생존자합을구해주세요


```python
survival_counts = titanic_df.groupby('Sex')['Survived'].sum()

print(survival_counts)
```

# 3) 클래스(Pclass)별생존자합을구해주세요


```python
survival_counts_by_class = titanic_df.groupby('Pclass')['Survived'].sum()

print(survival_counts_by_class)
```

# 4) 나이(Age)를활용유아,10대,20대,30대,40대,50대,60대,노인의생존자합을구해주세요.


```python

# 연령대(Age_Group)를 추가.
def categorize_age(age):
    if age < 10:
        return '유아 (0-9세)'
    elif age >= 10 and age < 20:
        return '10대 (10-19세)'
    elif age >= 20 and age < 30:
        return '20대 (20-29세)'
    elif age >= 30 and age < 40:
        return '30대 (30-39세)'
    elif age >= 40 and age < 50:
        return '40대 (40-49세)'
    elif age >= 50 and age < 60:
        return '50대 (50-59세)'
    elif age >= 60 and age < 70:
        return '60대 (60-69세)'
    else:
        return '노인 (70세 이상)'

titanic_df['Age_Group'] = titanic_df['Age'].apply(categorize_age)

# 연령대(Age_Group)별 생존자 수 구하기
survival_counts_by_age_group = titanic_df.groupby('Age_Group')['Survived'].sum()

print(survival_counts_by_age_group)
```
